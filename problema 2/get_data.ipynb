# Descargar dataset UTKFace

Este notebook descarga el dataset de Kaggle y copia las imágenes a la carpeta `UTKFace/`
al lado de los notebooks, lista para usar en VS Code en Windows.

**Ejecuta este notebook UNA SOLA VEZ antes de correr `P2_01_dataloaders.ipynb`.**


```python
# Instalar kagglehub si no está instalado
%pip install kagglehub
```

```python
import kagglehub
import os
import shutil
from pathlib import Path

# Descarga el dataset (se guarda en el cache local de kagglehub)
# En Windows la ruta del cache es: C:/Users/<tu_usuario>/.cache/kagglehub/...
print('Descargando dataset UTKFace...')
path = kagglehub.dataset_download('jangedoo/utkface-new')
print(f'Dataset descargado en: {path}')
```

```python
# Explorar qué carpetas trajo el dataset
print('Contenido del dataset descargado:')
for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    if os.path.isdir(folder_path):
        images = [f for f in os.listdir(folder_path)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        print(f'  {folder}: {len(images):,} imágenes')
```

```python
# Copiar las imágenes de UTKFace a una carpeta local junto a los notebooks
# Así P2_01_dataloaders.ipynb puede usar simplemente Path('UTKFace')

SRC_DIR  = Path(path) / 'UTKFace'   # carpeta dentro del cache de kagglehub
DEST_DIR = Path('UTKFace')          # carpeta local junto a los notebooks

if not SRC_DIR.exists():
    raise FileNotFoundError(
        f'No se encontró la carpeta UTKFace en {path}\n'
        f'Carpetas disponibles: {os.listdir(path)}'
    )

DEST_DIR.mkdir(exist_ok=True)

EXTENSIONS = {'.jpg', '.jpeg', '.png'}
images = [p for p in SRC_DIR.iterdir()
          if p.suffix.lower() in EXTENSIONS]

print(f'Copiando {len(images):,} imágenes a {DEST_DIR.resolve()} ...')
print('Esto puede tardar un par de minutos...')

for i, src in enumerate(images):
    shutil.copy2(src, DEST_DIR / src.name)
    if (i + 1) % 2000 == 0:
        print(f'  {i + 1:,} / {len(images):,} copiadas...')

total = sum(1 for f in DEST_DIR.iterdir() if f.suffix.lower() in EXTENSIONS)
print(f'\nListo! {total:,} imágenes en ./{DEST_DIR}/')
```

```python
# Verificación final
sample = [f.name for f in Path('UTKFace').iterdir()][:5]
print('Muestra de archivos en UTKFace/:')
for name in sample:
    print(f'  {name}')
print('\nTodo listo. Ejecuta ahora P2_01_dataloaders.ipynb')
```
